```python
from pyecharts import Page,Bar
df = df[df.score.notnull()] # 去掉评分为空的值
for coun in country:
    data = df[df.regions==coun]
    pscore = []
    type = []
    for key in keys:
        score = []
        for i in data.index:
            tp = data.loc[i, "types"].split('/')
            if key in tp:
                score.append(df.loc[i, "score"])
        if len(score)<=0:
            continue
        pscore.append(round(sum(score)/len(score), 1)) # round 保留一位小数
        type.append(key)
    bar = Bar("")
    bar.add("各个国家类别平均评分", type, pscore, mark_point=['min','max'],yaxis_max=10)
    path = "regions_type_html/"+coun+".html"
    bar.render(path)
```
