```python
for i in range(len(df.title)):
    df.loc[i, 'title'] = re.sub(r"\\n", "",df.loc[i, 'title'])
    df.loc[i, 'title'] = re.sub(r"\['", "",df.loc[i, 'title'])
    df.loc[i, 'title'] = re.sub(r"']", "", df.loc[i, 'title'])
    df.loc[i, 'title'] = re.sub(r"\(豆瓣\)", "",df.loc[i, 'title'])
    df.loc[i, 'title'] = re.sub(r" ", "", df.loc[i, 'title'])
```

```python
for i in range(len(df.text)):
    df.loc[i, 'text'] = re.sub(r"\\u3000\\u3000", "", df.loc[i, 'text'])
    df.loc[i, 'text']  = re.sub(r"\\n", "", df.loc[i, 'text'])
    df.loc[i, 'text']  = re.sub(r"<br />", "", df.loc[i, 'text'])
    df.loc[i, 'text']  = re.sub(r"\['", "", df.loc[i, 'text'])
    df.loc[i, 'text']  = re.sub(r"']", "", df.loc[i, 'text'])
    df.loc[i, 'text']  = re.sub(r"\[", "", df.loc[i, 'text'])
    df.loc[i, 'text']  = re.sub(r"\]", "",  df.loc[i, 'text'])
    df.loc[i, 'text']  = re.sub(r" ", "", df.loc[i, 'text'])
```

```python
for i in range(len(df.year)):
    df.loc[i, 'year'] = re.sub("']", "", df.loc[i, 'year'])
    df.loc[i, 'year'] = re.sub("\['", "", df.loc[i, 'year'])
    df.loc[i, 'year'] = re.sub("\]", "", df.loc[i, 'year'])
    df.loc[i, 'year'] = re.sub("\[", "", df.loc[i, 'year'])
```

```python
for i in range(len(df.director)):
    df.loc[i, 'director'] = re.sub("']", "", df.loc[i, 'director'])
    df.loc[i, 'director'] = re.sub("\['", "", df.loc[i, 'director'])
    df.loc[i, 'director'] = re.sub("\]", "", df.loc[i, 'director'])
    df.loc[i, 'director'] = re.sub("\[", "", df.loc[i, 'director'])
    df.loc[i, 'director'] = re.sub("', '", "/", df.loc[i, 'director'])
```

```python
for i in range(len(df.attrs)):
    df.loc[i, 'attrs'] = re.sub("']", "", df.loc[i, 'attrs'])
    df.loc[i, 'attrs'] = re.sub("\['", "", df.loc[i, 'attrs'])
    df.loc[i, 'attrs'] = re.sub("\]", "", df.loc[i, 'attrs'])
    df.loc[i, 'attrs'] = re.sub("\[", "", df.loc[i, 'attrs'])
    df.loc[i, 'attrs'] = re.sub("', '", "/", df.loc[i, 'attrs'])
```

```python
for i in range(len(df.types)):
    df.loc[i, 'types'] = re.sub("']", "", df.loc[i, 'types'])
    df.loc[i, 'types'] = re.sub("\['", "", df.loc[i, 'types'])
    df.loc[i, 'types'] = re.sub("\]", "", df.loc[i, 'types'])
    df.loc[i, 'types'] = re.sub("\[", "", df.loc[i, 'types'])
    df.loc[i, 'types'] = re.sub("', '", "/", df.loc[i, 'types'])
```

```python
for i in range(len(df.regions)):
    df.loc[i, 'regions'] = re.sub("']", "", df.loc[i, 'regions'])
    df.loc[i, 'regions'] = re.sub("\['", "",df.loc[i, 'regions'])
    df.loc[i, 'regions'] = re.sub("\]", "", df.loc[i, 'regions'])
    df.loc[i, 'regions'] = re.sub("\[", "", df.loc[i, 'regions'])
    df.loc[i, 'regions'] = re.sub(" / ", "/", df.loc[i, 'regions'])
    df.loc[i, 'regions'] = re.sub("', '", "/", df.loc[i, 'regions'])
```

```python
for i in range(len(df.language)):
    df.loc[i, 'language'] = re.sub("']", "", df.loc[i, 'language'])
    df.loc[i, 'language'] = re.sub("\['", "", df.loc[i, 'language'])
    df.loc[i, 'language'] = re.sub("\]", "", df.loc[i, 'language'])
    df.loc[i, 'language'] = re.sub("\[", "", df.loc[i, 'language'])
    df.loc[i, 'language'] = re.sub(" / ", "/", df.loc[i, 'language'])
    df.loc[i, 'language'] = df.loc[i, 'language'].strip()
```

```python
for i in range(len(df.release_date)):
    df.loc[i, 'release_date'] = re.sub("']", "",df.loc[i, 'release_date'])
    df.loc[i, 'release_date'] = re.sub("\['", "", df.loc[i, 'release_date'])
    df.loc[i, 'release_date'] = re.sub("\]", "", df.loc[i, 'release_date'])
    df.loc[i, 'release_date'] = re.sub("\[", "", df.loc[i, 'release_date'])
    df.loc[i, 'release_date'] = re.sub(" / ", "/", df.loc[i, 'release_date'])
    df.loc[i, 'release_date'] = re.sub(" ", "/", df.loc[i, 'release_date'])
```

```python
for i in range(len(df.runtime)):
#     df.loc[i, 'runtime'] = re.sub("']", "", df.loc[i, 'runtime'])
#     df.loc[i, 'runtime'] = re.sub("\['", "", df.loc[i, 'runtime'])
#     df.loc[i, 'runtime'] = re.sub("\]", "", df.loc[i, 'runtime'])
#     df.loc[i, 'runtime'] = re.sub("\[", "", df.loc[i, 'runtime'])
    df.loc[i, 'runtime'] = re.sub(" ", "", df.loc[i, 'runtime'])
    try:
        df.loc[i, 'runtime'] = re.findall(r"(\d+分钟)",df.loc[i, 'runtime'])[0]
    except:
        df.loc[i, 'runtime'] = ""
    finally:
        df.loc[i, 'runtime'] = re.sub(r"[^\d]*", "", df.loc[i, 'runtime'])
```

```python
for i in range(len(df.rating_people)):
    df.loc[i, 'rating_people'] = re.sub("']", "", df.loc[i, 'rating_people'])
    df.loc[i, 'rating_people'] = re.sub("\['", "", df.loc[i, 'rating_people'])
    df.loc[i, 'rating_people'] = re.sub("\]", "", df.loc[i, 'rating_people'])
    df.loc[i, 'rating_people'] = re.sub("\[", "", df.loc[i, 'rating_people'])
    df.loc[i, 'rating_people'] = re.sub(" ", "", df.loc[i, 'rating_people'])
    
```

```python
for i in range(len(df.score)):
    df.loc[i, 'score'] = re.sub("']", "", df.loc[i, 'score'])
    df.loc[i, 'score'] = re.sub("\['", "", df.loc[i, 'score'])
    df.loc[i, 'score'] = re.sub("\]", "", df.loc[i, 'score'])
    df.loc[i, 'score'] = re.sub("\[", "",df.loc[i, 'score'])
```
